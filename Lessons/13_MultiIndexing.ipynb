## Creating Your Own MultiIndex

```python
pops = pd.read_csv("data/state_pops.csv")
pops
```

```python
pops.index
```

```python
pops.set_index("state") #state as our index is not unique
#we want some unique labels for every row
```

```python
pops.set_index("year")
```

```python
pops.set_index(["state", "year"]) #order matters here
#state will be at zero level
#then year will be work as index in level 1
```

```python
pops.set_index(["year", "state"])
```

```python
pops.set_index(["state", "year"], inplace=True)
```

```python
pops
```

## Sorting A MultiIndex

```python
pops.sort_index()
#it will consider both state & year to sort rows =)
```

```python
pops.sort_index().head(20)
```

```python
pops.sort_index(ascending=False)
```

```python
#level0 is state
#level1 is year
pops.sort_index(level=1)
#sorts by year
```

```python
pops.sort_index(ascending=[False,True])
#pops.sort_index(level=[0,1],ascending=[False,True])
```

```python
pops.sort_index(level=[1,0],ascending=[False,True])
#asending for state
#descending for year
```

```python
pops.sort_index(inplace=True)
```

```python
pops
```

## get_level_values( )

```python
pops.index.levels
#gives all the different values for different levels
#here we have 2sublists
#fist is the list of all states or first level
#second is the list of all years or second level
```

```python
pops.index.get_level_values(0)
#get all different values for a given level of index
```

```python
pops.index.get_level_values(0)[10]
```

```python
pops.index.get_level_values(1)
```

```python
pops.loc[:, [1990, 1992, 1994], :]
```

```python
even_years = pops.index.get_level_values(1) % 2 == 0 
even_years
```

```python
len(even_years)
```

```python
len(pops)
```

```python
pops[even_years]
```

```python
pops[pops["population"] % 2 == 0]
```

```python
even_pops = pops["population"] % 2 == 0
```

```python
pops[even_years & even_pops]
```

```python
ends_with_a = pops.index.get_level_values(0).str[1] == "A"
#compare second character of each strings with "A"
ends_with_a
```

```python
pops[ends_with_a]
```

```python
pops[ends_with_a].index.values
```

## Stack( ) and Unstack( )

```python
pops
```

```python
pops.unstack()
#we have column for evey single year
```

```python
unstacked_df = pops.unstack(level="state")
unstacked_df
```

```python
unstacked_df.stack().
#back to having multi index
```

```python
unstacked_df.stack().unstack()
```

```python
titanic.groupby(["pclass", "sex"])["age"].mean()
```

```python
titanic.groupby(["pclass", "sex"])["age"].mean().plot(kind="bar")
```

```python
titanic.groupby(["pclass", "sex"])["age"].mean().unstack()
```

```python
titanic.groupby(["pclass", "sex"])["age"].mean().unstack().plot(kind="bar")
#draw 2bars per class
```

```python
titanic.groupby(["pclass", "sex"])["age"].mean().unstack(level="pclass")
```

```python
titanic.groupby(["pclass", "sex"])["age"].mean().unstack(level="pclass").plot(kind="bar")
```

```python
titanic.groupby(["sex", "survived"])["age"].mean()
```

```python
titanic.groupby(["sex", "survived"])["age"].mean().unstack()
```

```python
titanic.groupby(["sex", "survived"])["age"].mean().unstack().plot(kind="bar")
```

```python
df = titanic.groupby(["sex", "survived"])["age"].mean()
df.unstack().rename(columns={0: "Died", 1: "Survived"}).plot(kind="bar")
```
