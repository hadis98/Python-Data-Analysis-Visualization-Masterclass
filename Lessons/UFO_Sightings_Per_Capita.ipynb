```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ufos = pd.read_csv("data/nuforc_reports.csv", parse_dates=["date_time"])
```

```python
ufos.state.value_counts()
```

```python
populations = pd.read_csv("data/nst-est2020.csv")
```

```python
ufos
```

```python
pops = populations[["NAME", "POPESTIMATE2019"]]
```

```python
state_names = pd.read_csv("data/states.csv")
```

```python
state_names
```

```python
pops
```

```python
pops.merge(state_names, left_on="NAME", right_on="State").set_index("Code")["POPESTIMATE2019"]
```

```python
pop_series = pops.merge(state_names, left_on="NAME", right_on="State").set_index("Code")["POPESTIMATE2019"]
```

```python
ufos_2019 = ufos[ufos.date_time.dt.year == 2019]
```

```python
ufos_2019.state.value_counts()
```

```python
df = pd.concat([ufos_2019.state.value_counts(), pop_series], axis=1, join="inner")
```

```python
df.rename({"state": "num_sightings", "POPESTIMATE2019": "pop"}, axis=1, inplace=True)
```

```python
df
```

```python
df["sightings_per_cap"] = df["num_sightings"] / df["pop"]
```

```python
df["sightings_per_cap"].sort_values(ascending=False).head(10)
```

```python
plt.style.use("fivethirtyeight")
```

```python
df["sightings_per_cap"].sort_values(ascending=False).head(10).plot(kind="barh", title="2019 Sightings Per Capita")
```
