## loc[] & iloc[]

```python
countries.loc["Yemen"]
#it reaturns a series
```

```python
countries.loc[["Yemen"]]
#it reaturns a dataframe
```

```python
countries.loc[["Iran"]]
```

```python
countries.loc[["Canada"]]
```

```python
countries.loc[["Canada", "Mexico", "United States"]]
```

```python
titanic.loc[7]
#we get a series back
```

```python
titanic.loc[[7,9,876]]
```

```python
titanic.loc[5:10]
#use slices to get rows
```

```python
titanic.loc[5:10:2]
#2 is our step, 5+2,7+2,..
```

```python
countries.sort_index(inplace=True)
```

```python
countries
```

```python
countries.loc["Denmark":"France"]
```

```python
countries.loc["Denmark":"France":2]
```

```python
countries
```

```python
countries.iloc[0]
#first row of countries dataframe
```

```python
countries.iloc[[0]]
#get dataframe instead of series
```

```python
countries.iloc[[10,20,30,100]]
```

```python
countries.iloc[50:60]
```

```python
houses
```

```python
houses.sort_index(ascending=False, inplace=True)
```

```python
houses
```

```python
houses.iloc[0:5]
#iloc looks for position in dataframe not actual label
#it is based on the position of row not index number
#but .loc cares about the actual label
```

```python
houses.loc[21612:21611]
```

```python
houses.loc[21612:21611, ['price', 'bedrooms']]
```

```python
titanic.loc[50:60:2, ['name', 'sex', 'age']]
```

```python
countries.loc["Canada": "Denmark", ['Ladder score']]
```

```python
houses["bedrooms"].value_counts().iloc[0] # in position 0 => 3:9824 works as a index in array
```

```python
houses["bedrooms"].value_counts().loc[0] # 0:13 works as a key,value
```
