# Filtering Data Exercise
Answer the following questions using the `bestsellers.csv` dataset

```python
#Find the books written by Pete Souza
```

```python
import pandas as pd
```

```python
books = pd.read_csv('data/bestsellers.csv')
books
```

```python
books[books["Author"] == "Pete Souza"]
```

```python
# Find the books that are under 10 dollars
```

```python
books[books["Price"] < 10]
```

```python
#Find the books that have a price between 50 and 60 dollars
```

```python
books_greater_50 = books["Price"] > 50
books_less_60 = books["Price"] < 60
books[books_greater_50 & books_less_60]
```

```python
books[books["Price"].between(50,60)]
```

```python
#Find all the books written by Kristin Hannah, Andy Weir, or Delia Owens
```

```python
books[books["Author"].isin(["Kristin Hannah", "Andy Weir","Delia Owens"])]
```

```python
#Find the Non Fiction books that are rated 4.9
```

```python
nonfiction_books = books["Genre"] == "Non Fiction"
selected_rate = books["User Rating"] == 4.9
books[nonfiction_books & selected_rate]
```

```python
#Find the fiction book with the lowest User Rating
```

```python
fiction_books = ~nonfiction_books
lowest_rating = books["User Rating"] == books["User Rating"].min()
books[fiction_books & lowest_rating]
```

```python
books[books["Genre"] == "Fiction"].sort_values("User Rating")
```

```python
#Find 2012's top 5 Fiction books with the most Reviews
```

```python
year_2012 = books["Year"] == 2012
selected_books = books[fiction_books & year_2012]
selected_books.sort_values("Reviews",ascending=False).head()
```

```python
#Create a bar plot showing the 5 authors who have the most books with a rating under 4.5
```

```python
low_rated = books["User Rating"] < 4.5
#authors = books[low_rated]["Author"]
authors = books[low_rated]["Author"].value_counts().head()
authors.plot(kind='bar')
```
